```python
%matplotlib inline
import numpy as np
import pandas as pd

df = pd.read_csv("data_assignment.csv")
df.head()
```

```python
# Counting the number of job posts per city

pd.set_option('display.max_rows', None)
df1 = df.drop(['Id','Date','Company','Area','SubClassification','Requirement','FullDescription','Classification','LowestSalary','HighestSalary'], axis=1)
df1.groupby(['Location']).count().sort_values('Title',ascending=False)
```

```python
#Counting the number of job posts per city and type

df1.groupby(['Location','JobType']).count()
```

```python
#Counting the number of job posts per city and classification/sector and sorting
df2 = df.drop(['Id','Date','Company','Area','SubClassification','Requirement','FullDescription','JobType','LowestSalary','HighestSalary'], axis=1)
df2.groupby(['Location','Classification']).count().sort_values('Title',ascending=False)
```

```python
#Test
df3 = df.drop(['Id','Date','Company','Area','SubClassification','Requirement','FullDescription','JobType','LowestSalary','Classification'], axis=1)
df3.groupby(['Location'])['Title'].count().reset_index()
```
